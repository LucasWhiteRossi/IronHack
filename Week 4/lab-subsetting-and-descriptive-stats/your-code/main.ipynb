#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

```python
# your answer here
min_sal = employee.groupby(by='Department').min()._get_numeric_data()
mean_sal = employee.groupby(by='Department').mean()._get_numeric_data()
max_sal = employee.groupby(by='Department').max()._get_numeric_data()

print(min_sal)
print(mean_sal)
print(max_sal)
```
