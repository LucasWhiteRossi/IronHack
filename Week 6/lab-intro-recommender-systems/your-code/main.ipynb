```python
def recom_metric(metric, data):
    agr = data.groupby(by=['CustomerID','ProductName'])['Quantity'].sum()
    agr = pd.DataFrame(agr)
    product_by_customer = pd.pivot_table(data =agr, columns='CustomerID', index='ProductName', values='Quantity', fill_value=0)
    similar = pd.DataFrame(1/(1+squareform(pdist(product_by_customer.T, metric))), 
                            index=product_by_customer.columns, 
                            columns=product_by_customer.columns)
    recomendations = {}
    clients = list(set(data.CustomerID))

    def recomending_stuff(client):
        list_similar = similar[client].sort_values(ascending=False).head(6).index[1:] #step4
        step5 = agr.loc[list_similar] #step5
        step6 = step5.groupby(by='ProductName')["Quantity"].sum().sort_values(ascending=False) #step6
        customer_list = agr.loc[client]
        merged = pd.merge(customer_list,step6,on='ProductName', how='outer').fillna(0)
        merged.columns=['customer', 'similar_others']
        results = list(merged[(merged['customer']==0)&(merged['similar_others']!=0)]['similar_others'].sort_values(ascending=False).head().index)
        return results

    for client in clients:
        recomendations[client] = recomending_stuff(client)

    return recomendations
```
