####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url7 = 'https://www.emsc-csem.org/Earthquake/'
````

````python
page = requests.get(url7)
````

````python
page.status_code
````

````python
page.content[0:700]
````

````python
soup = BeautifulSoup(page.content, 'html.parser')
````

````python
dates = soup.find_all('a')
dates
````

````python
dates2 = [date.get_text() for date in dates if "Earthquake/earthquake.php?id" in str(date)]
dates2
````

````python
# Results for dates and time
dates3 = [date.split()[0] for date in dates2]
time = [date.split()[1] for date in dates2]
````

````python
lat_long = soup.find_all('td', class_='tabev1')
lat_long[0:5]
````

````python
lat_long[0].get_text().strip()
````

````python
lat_long2 = [item.get_text().strip() for item in lat_long]
lat_long2[0:4]
````

````python
#indices
ind = [num for num in range(len(lat_long2))]

#filtering latitude and longitude
lat = [lat_long2[num] for num in ind if num%2==0]
long = [lat_long2[num] for num in ind if num%2!=0]
````

````python
#to find directions
direc = soup.find_all('td', class_='tabev2')

direc2 = [arrow.get_text().split()[0] for arrow in direc if arrow.get_text().split()[0]in ['N','W',"S", "E"]]

direc2[0:10]
````

````python
#indices
ind = [num for num in range(len(direc2))]

#filtering latitude and longitude directions
direc_lat = [direc2[num] for num in ind if num%2==0]
direc_long = [direc2[num] for num in ind if num%2!=0]
````

````python
direc_lat[0:10]
````

````python
direc_long[0:10]
````

````python
#region name
````

````python
region = soup.find_all('td', class_='tb_region')
region[0:4]
````

````python
region2 = [reg.get_text().strip() for reg in region] 
region2
````

````python
table = pd.DataFrame({"DATA":dates3, "Horario": time, "Latitude": lat, "Dir_Lat": direc_lat, "Longitude": long, "Dir_Long": direc_long, "Region": region2})

table.head(20)
````

### Count number of tweets by a given Twitter account. You will need to include a try/except block for account names not found.
Hint: the program should count the number of tweets for any provided account

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/' 
````

````python
#let's try with Obama
url = 'https://twitter.com/barackobama' 
data = requests.get(url)
data.status_code
````

````python
soup = BeautifulSoup(data.content, 'html.parser')
````

````python
numbers = soup.find_all('span', class_="ProfileNav-value")
````

````python
n_tweets = ' '.join(numbers[0].get_text().split())
n_tweets
````

````python
def tweet_numbers(credential):
    url = 'https://twitter.com/'+str(credential)
    data = requests.get(url)
    if data.status_code != 200:
        return "Error. Try to conect or verify the account"
    else:
        soup = BeautifulSoup(data.content, 'html.parser')
        try:
            numbers = soup.find_all('span', class_="ProfileNav-value")
            n_tweets = ' '.join(numbers[0].get_text().split())
            return n_tweets
        except:
            return "Not a valid entry"
        
````

````python
tweet_numbers("realdonaldtrump")
````

### List all language names and number of related articles in the order they appear in wikipedia.org


````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
data = requests.get(url)
data.status_code
````

````python
soup = BeautifulSoup(data.content, 'html.parser')
````

````python
lang = soup.find_all("a", class_="link-box")
#lang[0]
#lang[0]['title'].split("")[0]
languages = [lan['title'].split("")[0] for lan in lang]
articles = [re.findall(r'<bdi dir=\"ltr\">(.*)</bdi>', str(lan))[0] for lan in lang]

wiki = pd.DataFrame({"Languages": languages, "articles number": articles})

wiki
````
