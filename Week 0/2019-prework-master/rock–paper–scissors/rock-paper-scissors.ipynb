```python
# Import the choice function of the random module
from random import choice

# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

Options = ['stone','paper','scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...

MaxNumGames = 5

#__________________________________________________________________________
# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games

MaxToWin = 3

#___________________________________________________________________________
# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

def jokenpo(Options):
    return choice(Options)


#___________________________________________________________________________
# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

def game():

    condicao = True
    while condicao:
        YourTurn = input("Choose stone, paper or scissors")
        YourTurn = YourTurn.lower()
        if YourTurn in Options:
            condicao = False
        else:
            print("\nSomething went wrong\n")
    

    
    MachineTurn = jokenpo(Options)
    return (YourTurn, MachineTurn)

#_________________________________________________________________________________________________
# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins

def combat(YourTurn, MachineTurn):
    
    human_wins = [["scissors","paper"], ["paper", "stone"], ["stone", "scissors"]]
    tie = [["scissors","scissors"], ["paper","paper"], ["stone", "stone"]]
    machine_wins = [["paper","scissors"], ["stone","paper"], ["scissors","stone"]]

    
    if ([YourTurn, MachineTurn] in human_wins):
        return 2
    if ([YourTurn, MachineTurn] in machine_wins):
        return 1
    if ([YourTurn, MachineTurn] in tie):
        return 0

#_________________________________________________________________________________  
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated

def show_status(HumanWins, MachineWins):
    YourTurn, MachineTurn = game()
    result = combat(YourTurn, MachineTurn)
    
    message = "Error"
    if result == 2:
        message = "You win!"
        HumanWins += 1
    if result == 1:
        message = "You loose!"
        MachineWins += 1
    if result == 0:
        message = "Tie!"
    
    print("\n*********************************************************")
    print("[You: {} ] VS. [Machine: {}]\nResult of the turn: {}".format(YourTurn, MachineTurn, message))
    print("Points")
    print("You: ", HumanWins , " | Machine: ", MachineWins, "")
    print("*********************************************************")
    print("\n____________________________________________________")
    return (HumanWins, MachineWins)
    
# Create two variables that accumulate the wins of each participant

HumanWins = 0
MachineWins = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.

condicao2 = True
counter = 0
while condicao2:
    HumanWins, MachineWins = show_status(HumanWins, MachineWins)
    counter += 1
    if counter >= MaxNumGames:
        condicao2 = False
        break
    if HumanWins >= MaxToWin:
        condicao2 = False
        break
    if MachineWins >= MaxToWin:
        condicao2 = False
        break
    
# Print by console the winner of the game based on who has more accumulated wins

if (HumanWins > MachineWins):
    print("\nVictory! You win the game.\n--------------------")
elif (HumanWins == MachineWins):
    print("\nTie! Try again.\n--------------------")
else:
    print("\nsorry... you loose the game.\n--------------------")
    
```
